```python
import torch

# Create a 1D tensor of size 3
x = torch.tensor([10, 20, 30])  # Shape: [3]

# Add a fake dimension to make it 2D
x = x.unsqueeze(0)              # Shape: [1, 3]

# Apply expand
y = x.expand(1, -1)             # Shape: [1, 3]
print(y)
```

```python
ones = torch.ones((1, 256))
zeros = torch.zeros((1, 256))

attention_mask = torch.cat([ones, zeros], dim=1)
print(attention_mask)
```

```python
torch.all(attention_mask == 1)
```
